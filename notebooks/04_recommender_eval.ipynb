# 04 Recommender Evaluation

Lightweight qualitative and coverage checks for a content-based recommender.

```python
import sys
from pathlib import Path

_nb = Path("notebooks/_project.py")
sys.path.insert(0, str(_nb.parent if _nb.exists() else Path("_project.py").parent))
from _project import setup

setup()

from app.services.ml.tfidf_recommender import MovieRecommender

recommender = MovieRecommender.load('artifacts')
queries = ['The Matrix', 'Inception', 'Interstellar']
for q in queries:
    print('\nQUERY:', q)
    for r in recommender.recommend(q, k=5, mode='title'):
        print(' -', r['title'], r['score'])
```

```python
# Catalog coverage
catalog = recommender.catalog_
print('Movies:', len(catalog))
print('Unique genres:', len(set(g for genres in catalog['genres'] for g in genres)))
print('TF-IDF matrix:', recommender.matrix_.shape)
```
